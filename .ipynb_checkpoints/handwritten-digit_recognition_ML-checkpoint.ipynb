```python
#potrebne knjižnice 
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.optimizers import SGD
from tensorflow.compat.v1.ragged import RaggedTensorValue
from tensorflow.compat.v1 import executing_eagerly_outside_functions
from tensorflow.keras.models import load_model
import cv2
```

```python
mnist = tf.keras.datasets.mnist # spremanje slika iz dataseta u varijablu mnist
(x_train,y_train),(x_test,y_test) = mnist.load_data() #punjenje podataka za treniranje i testiranje
print(x_train.shape) # izgled podataka za treniranje 60000 slika velicine 28 x 28
print(x_test.shape) # testiranje 10000 slika od 28 * 28
print(y_train[0]) #to je label i on nam pokazuje za svaku sliku koji bi broj trebao biti na njoj
#također postoji 60000 slika za to kao i za treniranje
```

```python
#provjera kako izgledaju podaci na kojima cemo testirati model
plt.figure()
plt.imshow(x_train[0])#naredba za prikaz slike
plt.title("Pravi izgled slike")
plt.show()
#potrebno je sliku pretvoriti u binarni format jer neznamo tocno kojeg su 
#oblika sve slike koje se testiraju
plt.imshow(x_train[0], cmap = plt.cm.binary)
plt.title("Slika u binarnom formatu")
plt.show()
```

```python
#u RGB modelu boja crna boja je 0 a bijela je 255 tako da cemo sad prikazati vrijednosti svih piksela
print(x_train[0])
#slika od iznad koju smo plotali je invertana, zapravo je pozadina crna i sama brojka je bijela sto se vidi na ovom ispisu
#to se događa jer funckija plt.cm.binary koristi drugačije mapiranje
```

```python
#prije samog treniranja potrebno je ove brojke normalizirati jer se na gornjoj slici moze viditi kako osim samih
#vrijednosti 255 imamo i druge koje su blizu , to su varijacije sive boje
x_train = tf.keras.utils.normalize(x_train, axis = 1)
x_test = tf.keras.utils.normalize(x_test, axis = 1)
plt.imshow(x_train[0], cmap = plt.cm.binary)
```

```python
#sada kada je to normalizirano sve su vrijednosti između 0 i 1 a ne vise izmedu 255
print(x_train[0])
```

```python
#sada kada to sve imamo moramo dodati slikama jos jednu dimaneziju jer ce nam trebati kod konvolucijskog sloja
IMG_SIZE = 28 #globalna varijabla veličine slike
x_train_added_dim = np.array(x_train).reshape(-1, IMG_SIZE, IMG_SIZE, 1) # dodatna dimenzija predstavlja kernel filter
x_test_added_dim = np.array(x_test).reshape(-1, IMG_SIZE, IMG_SIZE, 1) #-1 predstavlja veličinu arraya tj. cijeli array
print(x_train_added_dim.shape)
print(x_test_added_dim.shape)
#evo sada imam0 sve slike za testiranje i treniranje pretvorene iz 2D slika u 3D slike
```

```python
#sada kada sve ovo imamo mozemo kreirati arhitekturu dubokih mreža (Deep learning)
model = Sequential() # ovo predstavlja da je model napravljen na nacin da se slojevi izvode redoslijedom

#PRVI KONVOLUCIJSKI SLOJ
model.add(Conv2D(64, (3,3), input_shape = x_train_added_dim.shape[1:])) #64 označava broj filtera gdje je svaki 3x3
#ide se od indeksa 1 jer je prvi 60000 a treba nam 28,28,1 i to se uzima u filter
model.add(Activation("relu")) # ova funckija mice sve sta je <0 ako se dogodi a sve sta je >0 to uzima u novi layer
model.add(MaxPooling2D(pool_size=(2,2))) #služi za izdvajanje ključnih značajki
#uzimaju se blokovi po 2x2 i uzima samo najveca vrijednost 

#DRUGI KONVOLUCIJSKI SLOJ
model.add(Conv2D(64, (3,3)))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

#TRI KONVOLUCIJSKA SLOJA
model.add(Conv2D(64, (3,3)))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

#Svaki naredni konvolucijski sloj prima ključne podatke prethodnog
#Svaki naredni sloj se smanjuje tako da prvi konvolucijski sloj ima 26x26 slike
#Drugi 24x24 i tako dalje a MaxPooling ih smanjuje na pola znaci 26x26 na 13x13
#Nakon tri konvolucijska sloja ide sloj za povezivanje

#PRVI POVEZANI SLOJ
model.add(Flatten()) # Ovo služi kako bi se te slike tipa 24x24 = 576 znaci da se iz 2D pretvori nazad u 1D
model.add(Dense(64)) # potpuno povezan sloj sa 64 neurona u neuronskoj mrezi
model.add(Activation("relu")) #relu je parametar aktivacijske funckije koji sve negativne ulaze vraca kao 0 a pozitivne kao sto jesu

#DRUGI POVEZANI SLOJ
model.add(Dense(32))
model.add(Activation("relu"))

#potrebno je iz sloja u sloj doći do 10 neurona jer imamo 10 mogucih znamenki na ulazu

#POSLJEDNJI POVEZANI SLOJ
model.add(Dense(10))
model.add(Activation("softmax")) #posljednja aktivacijska funckija mora biti tipa softmax ili nesto slicno gdje se
#realni brojevi pretvaraju u vjerojatnosti koje su u zbroju jednake 1

#Ovdje je kompletan pregled modela gdje se vidi svaki definirani sloj te njegov izgled
model.summary()
```

```python
#Prije korištenja samog modela potrebno ga je kompajlirat
model.compile(loss = "sparse_categorical_crossentropy", optimizer = "adam", metrics = ['accuracy'])
```

```python
#Sada kada imamo kompajliran model moze se istrenirat
model.fit(x_train_added_dim, y_train, epochs = 5, validation_split = 0.3) # koristi se set kojem smo dodali trecu dimenziju
#te y_train kao label sto je sto 
#potrebno je provjeriti odnose između accuracy i val_accuracy
#postavljeno je da se 30% podataka koristi za validaciju te ukoliko su ta dva broja slicni znaci da je treniranje dobro
#ukoliko je velika razlika dolazi do takozvanog overfittinga gdje je mreza prenapucena te ima previse slojeva i treba je reducirat
```

```python
#kada je model istreniran potrebno je provesti test i vidjeti koliko se podataka izgubi te koliko je accuracy
test_loss, test_accuracy = model.evaluate(x_test_added_dim,y_test)
print(test_loss)
print(test_accuracy)
```

```python
#spremanje modela
# Nakon što završi treniranje modela
model.save("model_digit_recognition.keras")
```

```python
import tkinter as tk
from tkinter import Canvas, Button, Label
import cv2
import numpy as np
from PIL import ImageGrab
from tensorflow.keras.models import load_model

# Učitaj trenirani model
model = load_model('model_digit_recognition.keras')

class DigitRecognizerApp:
    def __init__(self, master):
        self.master = master
        self.master.title("Digit Recognizer")

        self.canvas = Canvas(master, width=200, height=200, bg="black", highlightthickness=0)
        self.canvas.pack()

        self.label = Label(master, text="Napisani broj:", fg="black")  # Početna boja teksta bijela
        self.label.pack()

        self.canvas.bind("<B1-Motion>", self.paint)
        self.canvas.bind("<ButtonRelease-1>", self.reset)

        self.last_x = None
        self.last_y = None
        self.current_color = "white"  # Promijenjena boja crteža u bijelu

        self.predict_button = Button(master, text="Predikcija", command=self.predict_digit)
        self.predict_button.pack()

        self.clear_button = Button(master, text="Očisti", command=self.clear_canvas)
        self.clear_button.pack()

    def paint(self, event):
        x, y = event.x, event.y
        if self.last_x is not None and self.last_y is not None:
            self.canvas.create_line(self.last_x, self.last_y, x, y, width=8, fill=self.current_color, capstyle=tk.ROUND, smooth=tk.TRUE)

        self.last_x = x
        self.last_y = y

    def reset(self, event):
        self.last_x = None
        self.last_y = None

    def predict_digit(self):
        # Spremi crtež kao sliku
        self.canvas.update()
        x = self.master.winfo_rootx() + self.canvas.winfo_x()
        y = self.master.winfo_rooty() + self.canvas.winfo_y()
        x1 = x + self.canvas.winfo_width()
        y1 = y + self.canvas.winfo_height()
        ImageGrab.grab().crop((x, y, x1, y1)).save("digit.png")
        
        test_img = cv2.imread('digit.png')
        gray = cv2.cvtColor(test_img, cv2.COLOR_BGR2GRAY)
        resized_img = cv2.resize(gray, (28,28), interpolation = cv2.INTER_AREA)
        resized_img_added_dim = tf.keras.utils.normalize(resized_img, axis = 1) # normalizacija slike 
        resized_img_added_dim = np.array(resized_img_added_dim).reshape(-1, IMG_SIZE, IMG_SIZE, 1)
        predictions = model.predict(resized_img_added_dim)
        predicted_class_index = np.argmax(predictions)
        print(predicted_class_index)
        # Ažuriraj oznaku na GUI-u
        self.label.config(text=f"Napisani broj: {predicted_class_index}", fg="black")

    def clear_canvas(self):
        self.canvas.delete("all")
        self.label.config(text="Napisani broj:", fg="black")

# Pokreni aplikaciju
root = tk.Tk()
app = DigitRecognizerApp(root)
root.mainloop()
```
